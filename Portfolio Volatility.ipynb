```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
# Function to extract asset tickers and their respective weights
def extract_tickers_and_weights(initial_multi_asset_portfolio, initial_gold_portfolio, initial_cash_portfolio):
    tickers = []
    invested_amounts = []
    
    # Extract data from multi-asset portfolio
    for asset_type, assets in initial_multi_asset_portfolio.items():
        for asset in assets:
            tickers.append(asset['ticker'])
            invested_amounts.append(asset['invested_amount'])
    
    # Extract data from gold portfolio
    for asset in initial_gold_portfolio:
        tickers.append(asset['ticker'])
        invested_amounts.append(asset['invested_amount'])
    
    # Sum total investment and calculate weights
    total_investment = sum(invested_amounts)
    weights = [amount / total_investment for amount in invested_amounts]
    
    return tickers, weights
```

```python
# Function to fetch historical data for all assets except cash (cash is treated as constant)
def fetch_historical_data(assets, start_date, end_date):
    data = {}
    
    for asset in assets:
        if asset != 'CASH':  # Skip fetching data for cash
            df = yf.download(asset, start=start_date, end=end_date)['Adj Close']
            data[asset] = df
        else:
            # Create a constant value series for cash
            cash_data = pd.Series([1] * len(pd.date_range(start=start_date, end=end_date)), 
                                  index=pd.date_range(start=start_date, end=end_date), 
                                  name='CASH')
            data[asset] = cash_data
    
    return pd.DataFrame(data)
```

```python
# Function to calculate daily returns of the assets (special handling for CASH)
def calculate_daily_returns(asset_data):
    daily_returns = asset_data.pct_change().fillna(0)  # Fill NaNs with 0 for no return (especially for cash)
    return daily_returns

# Function to calculate portfolio returns based on weights
def calculate_portfolio_returns(daily_returns, weights):
    portfolio_returns = daily_returns.dot(weights)
    return portfolio_returns

# Function to calculate the correlation matrix of the portfolio assets
def calculate_correlation(daily_returns):
    return daily_returns.corr()

# Function to calculate volatility (standard deviation) of portfolio returns
def calculate_portfolio_volatility(portfolio_returns):
    return np.std(portfolio_returns)
```

```python
def calculate_stability_score(avg_corr, portfolio_volatility):
    """
    Calculate the stability score of a portfolio on a scale from 0 to 10 based on its
    average correlation and volatility, using a more granular approach.

    Stability Score:
    - High correlation and high volatility indicate low stability.
    - High correlation but low volatility suggests higher stability.
    - Low correlation but high volatility suggests moderate stability.
    - Low correlation and low volatility indicate high stability.

    Parameters:
    avg_corr (float): Average correlation of assets in the portfolio.
    portfolio_volatility (float): Standard deviation (volatility) of the portfolio returns.

    Returns:
    float: Stability score between 0 and 10, where 0 is least stable and 10 is most stable.
    """
    # Define thresholds for correlation and volatility
    correlation_threshold = 0.75  # Threshold for high correlation
    volatility_threshold = 0.02   # Threshold for high volatility
    
    # Scale the average correlation and volatility to a 0-1 range relative to the thresholds
    scaled_corr = min(max(avg_corr / correlation_threshold, 0), 1)  # 0-1 scale
    scaled_volatility = min(max(portfolio_volatility / volatility_threshold, 0), 1)  # 0-1 scale

    # Calculate stability score:
    # High correlation increases risk, and high volatility increases risk.
    score = 10 * (1 - (0.5 * scaled_corr + 0.5 * scaled_volatility))

    # Ensure the score is within the 0-10 range
    return max(min(score, 10), 0)
```
